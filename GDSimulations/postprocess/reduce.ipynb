#### **Figure 2**

```python
path = '../run/Fig2/Simulators/'

Na = 5
lam = [0.5, 0.1, 0.05]

alpha = [[] for _ in lam]
MSE = [[] for _ in lam]
Loss = [[] for _ in lam]

for i in range(len(lam)):
    for l in range(Na):
        filename = path + 'SimGD_Fig2_0' + str(i) + str(l) + '.pickle'
        with open(filename, "rb") as file:
            S = pickle.load(file)
            alpha[i].append(S.alpha)
            MSE[i].append(S.MSE[:, -1].clone())
            Loss[i].append(S.loss[:, -1].clone())

alpha = [np.concatenate(a) for a in alpha]
MSE = [torch.cat(m) for m in MSE]
Loss = [torch.cat(l) for l in Loss]

alphaZ = []
Z = []

for i in range(len(lam)):
    filename = path + 'SimGD_Fig2_1' + str(i) + '.pickle'
    with open(filename, "rb") as file:
        S = pickle.load(file)
        alphaZ = S.alpha
        Z.append(torch.stack(S.Z))

torch.save(alpha, 'Fig2/alpha.pt')
torch.save(MSE, 'Fig2/MSE.pt')
torch.save(Loss, 'Fig2/loss.pt')
torch.save(alphaZ, 'Fig2/alphaZ.pt')
torch.save(Z, 'Fig2/Z.pt')
```

#### **Figure 12**

```python
path = '../run/Fig12/Simulators/'

data = ['hermite', 'gaussian']
lam = [1.0, 0.]
Na = 2

alpha = [[[] for _ in lam] for _ in data]
MSE = [[[] for _ in lam] for _ in data]
Loss = [[[] for _ in lam] for _ in data]
time = [[[] for _ in lam] for _ in data]

for i in range(len(data)):
    for j in range(len(lam)):
        for l in range(Na):
            filename = path + 'SimGD_Fig12_' + str(i) + str(j) + str(l) + '.pickle'
            with open(filename, "rb") as file:
                S = pickle.load(file)
                alpha[i][j].append(S.alpha)
                MSE[i][j].append(S.MSE)
                Loss[i][j].append(S.loss)
                time[i][j] = S.eta * S.steps_save

alpha = [[np.concatenate(a1) for a1 in a] for a in alpha]
MSE = [[torch.cat(m1) for m1 in m] for m in MSE]
Loss = [[torch.cat(m1) for m1 in m] for m in Loss]


torch.save(alpha, 'Fig12/alpha.pt')
torch.save(MSE, 'Fig12/MSE.pt')
torch.save(Loss, 'Fig12/loss.pt')
torch.save(time, 'Fig12/time.pt')
```
